### Resampling

```python
df_backup=df.copy()

wine_quality_0_cnt, wine_quality_1_cnt = df['wine quality'].value_counts()
print('wine quality_0 counts:',wine_quality_0_cnt, '\n wine quality_1 counts:',wine_quality_1_cnt)
```

```python
wine_quality_0_data = df[df['wine quality'] == 0]
wine_quality_1_data = df[df['wine quality'] == 1]

print('shape of wine quality_0:',wine_quality_0_data.shape)
print('shape of wine quality_1:',wine_quality_1_data.shape)
```

### Create Oversampling with some percentage of major class

```python
# Adding 90% data to class 1
wine_quality_1_oversamp_data = wine_quality_1_data.sample(n = round(wine_quality_0_cnt*0.9), random_state=10, replace=True)

df_sampling = pd.concat([wine_quality_1_oversamp_data, wine_quality_0_data], axis = 0)

print(df_sampling['wine quality'].value_counts())
```

```python
maxAccu=0
maxRs=0
for i in range(1,2000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=i)
    mod = DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best Accuracy is ", maxAccu, "On random_state", maxRS)
```
